```python
from PriorityQueue import PriorityQueue


def childNodes(i):
     return (2*i)+1, (2*i)+2

def printTree(a, i=0, d = 0):
    if (i == 0):
        print("---------------")
    if i >= len(a):
        return 
    l, r =  childNodes(i)
    printTree(a, r, d = d+1)
    print("   "*d + str(a[i]))
    printTree(a, l, d = d+1)


test = PriorityQueue()

test.insert(1)
printTree(test.heap)

test.insert(6)
printTree(test.heap)

test.insert(15)

printTree(test.heap)

test.insert(16)

printTree(test.heap)

test.insert(25)

printTree(test.heap)

test.insert(5)
printTree(test.heap)

test.insert(35)
printTree(test.heap)

test.insert(11)
printTree(test.heap)

test.insert(8)
printTree(test.heap)
```
